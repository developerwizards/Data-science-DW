# **1. Scalar and vectors**


*   **Scalar** is single number denoted as x
*   **Vector** is an array of scalars  x1, x2, x3, ..., xn, named as x
      * by default vector is a column vector
      * Note that here indexing begins from 1, unlike python where it begin from 0

```
import numpy as np 
x = np.array([[4],[5],[6]])
print(x)
print(x.shape)
```

# **2. Matrices and tensors**

```
#Matrix
x = np.array([[4,5,6,3],
              [2,3,4,7],
              [8,7,9,5]])
print(x)
print(x.shape)
```

```
X= x.reshape((2,6))
print(X)
print(X.shape)
```

```
#Row vector
X = x.reshape((1,-1))
print(X)
print(X.shape)
```

```
#Column Vector
X = x.reshape((-1, 1))
print(X)
print(X.shape)
```

```
#Tensor
T = np.ones((5,5,3))
print(T)
```

#**MATRIX MULTIPLICATION**

Two types of multiplication:-
1.   Element wise Multiplication/ Hadamard product
 

            *   Shape of both matrix should be same

           


2.   Dot Product


            *   Number of columns in 1st matrix must be equal to number of rows in 2nd matrix
            * further, if 1st Matrix is of shape (m,n) and 2nd Matrix is of shape (n,p), then the resulting matrix is of shape (m,p)
            * C(i,j) = Summation(A(i,k)*B(k,j)
            * A more useful method to think of a matrix multiplication  is as a **linear combination of columns of A weighted by column entries of B**

# **Inverse of a Matrix**

```
inverse_A = np.linalg.inv(A)
print(inverse_A)
print(np.dot(A,inverse_A))
```

```
#Pseudo inverse
A = np.array([[4,2],
              [6,3]])
print(np.linalg.det(A))
try:
  print(np.linalg.inv(A))
except np.linalg.LinAlgError as e:
  print(e)

print("###################Pseudo#################")
print(np.linalg.pinv(A))
```
