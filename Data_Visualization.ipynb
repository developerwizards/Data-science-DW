```python
#Importing Required library
from scipy.integrate import quad
import numpy as np

x = np.linspace(-3, 3, num=100)
constant = 1.0/np.sqrt(2*np.pi)
snd = constant*np.exp((-x**2)/2.0)

fig, ax = plt.subplots(figsize = (10,5))
ax.plot(x, snd)
ax.set_ylim(0)
ax.set_title('Normal distribution', size = 20)
ax.set_ylabel('Probability density', size = 20)
plt.show()
```

The graph above does not show you the probability of events but their probability density. To get the probability of an event within a given range we will need to integrate. Suppose we are interested in finding the probability of a random data point landing within 1 standard deviation of the mean, we need to integrate from -1 to 1. This can be done with SciPy.

```python
# Make a PDF for the normal distribution a function
def normalProbabilityDensity(x):
    constant = 1.0 / np.sqrt(2*np.pi)
    return(constant * np.exp((-x**2) / 2.0) )
# Integrate PDF from -1 to 1
result, _ = quad(normalProbabilityDensity, -1, 1, limit = 1000)
print(result)
```

```python
result, _ = quad(normalProbabilityDensity, -2, 2, limit = 1000)
print(result)
```

```python
result, _ = quad(normalProbabilityDensity, -3, 3, limit = 1000)
print(result)
```

```python
import pandas as pd
df = pd.read_csv('movies_metadata.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
titles = list(df.get('title'))
```

```python
print(titles)
```

```python
for title in titles:
  if type(title) == float:
    print(title)
```

```python
freq_titles = {}

for title in titles:
  if type(title) != float:
    length = len(title)

    if freq_titles.get(length) is None:
      freq_titles[length] = 1
    else:
      freq_titles[length] += 1
      
```

```python
freq_titles
```

```python
X = np.array(list(freq_titles.keys()))
Y = np.array(list(freq_titles.values()))
print(X,Y)
```

```python
plt.scatter(X, Y)
plt.xlabel("Length of Movie title")
plt.ylabel("Freq Count")
plt.show()
```
